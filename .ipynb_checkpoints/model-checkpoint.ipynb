# Steinbeck–Akerlof ABM (Full Research Version)
Agent-based model with asymmetric information, forced demand, vulnerability, and market power.


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dataclasses import dataclass
import ipywidgets as widgets
from IPython.display import display
import os

os.makedirs("figures", exist_ok=True)
os.makedirs("tables", exist_ok=True)

@dataclass
class Params:
    n_buyers:int=1000
    n_sellers:int=100
    periods:int=52

    info_asymmetry:float=0.7
    urgency_mean:float=0.7
    literacy_mean:float=0.4

    seller_opportunism:float=0.8
    price_discrimination:float=0.7

    institution_strength:float=0.0

class Car:
    def __init__(self):
        self.quality = np.random.rand()
        self.lemon = self.quality < 0.5

class Seller:
    def __init__(self,p):
        self.p=p
        self.cars=[Car() for _ in range(np.random.randint(2,6))]

    def price(self,b,c):
        base = 100*c.quality
        markup = 1 + self.p.price_discrimination*(1-b.lit)*b.urg
        if c.lemon:
            markup += self.p.seller_opportunism
        return base*markup

class Buyer:
    def __init__(self,p):
        self.p=p
        self.budget=np.random.uniform(50,200)
        self.urg=np.clip(np.random.normal(p.urgency_mean,0.2),0,1)
        self.lit=np.clip(np.random.normal(p.literacy_mean,0.2),0,1)

    def perceived_q(self,c):
        return np.clip(c.quality + np.random.normal(0,self.p.info_asymmetry),0,1)

def run(p):
    out=[]
    for t in range(p.periods):
        buyers=[Buyer(p) for _ in range(p.n_buyers)]
        sellers=[Seller(p) for _ in range(p.n_sellers)]

        trades=0
        lemons=0
        prices=[]

        for b in buyers:
            s=np.random.choice(sellers)
            if not s.cars: continue
            c=np.random.choice(s.cars)

            price=s.price(b,c)
            wtp=b.budget*b.perceived_q(c)

            if wtp>=price or b.urg>0.85:
                trades+=1
                prices.append(price)
                if c.lemon: lemons+=1
                s.cars.remove(c)

        out.append({
            "t":t,
            "lemon_share":lemons/trades if trades>0 else 0,
            "avg_price":np.mean(prices) if prices else 0
        })

    return pd.DataFrame(out)

def experiment(p):
    df=run(p)
    df.to_csv("tables/results.csv",index=False)

    plt.figure()
    plt.plot(df["t"],df["lemon_share"])
    plt.title("Lemon Share")
    plt.savefig("figures/lemon_share.png")
    plt.close()

    plt.figure()
    plt.plot(df["t"],df["avg_price"])
    plt.title("Average Price")
    plt.savefig("figures/avg_price.png")
    plt.close()

    return df

def ui(n_buyers,info_asym,urgency,inst):
    p=Params(n_buyers=n_buyers,
             info_asymmetry=info_asym,
             urgency_mean=urgency,
             institution_strength=inst)
    df=experiment(p)
    display(df.head())

widgets.interact(ui,
                 n_buyers=widgets.IntSlider(1000,100,5000,100),
                 info_asym=widgets.FloatSlider(0.7,0,1,0.1),
                 urgency=widgets.FloatSlider(0.7,0,1,0.1),
                 inst=widgets.FloatSlider(0.0,0,1,0.1))
```
